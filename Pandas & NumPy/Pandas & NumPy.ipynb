```python
#Exercise 2a
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Assignment_table.csv", names = ['left', 'right'])
dfl = pd.DataFrame(data['left'])
dfr = pd.DataFrame(data['right'])

dfl.hist('left')
dfr.hist('right')
plt.show()
```

```python
#Exercise 2b(3 clusters) 
from numpy import *
from numpy.random import *
from pylab import *
# import pandas as pd/
# import matplotlib.pyplot as plt
from scipy.cluster.vq import *

data = np.array(data)

centroids,_ = kmeans(data, 3)

index,_ = vq(data, centroids)
# plt.scatter(dfl, dfr)

# x = data.iloc[:, 0:1]

centroids,_ = kmeans(data, 3)
index,_ = vq(data, centroids)

plot(data[index==0,0], data[index==0,1], 'or', data[index==1,0], data[index==1,1], 'ob', data[index==2,0], data[index==2,1], 'om')
plot(centroids[:,0], centroids[:, 1], 'sg', markersize = 8)


show()
print("I think 3 would be the right number of cluster for the dataset")
```

```python
#Exercise 2b(4 clusters)

data = np.array(data)

centroids,_ = kmeans(data, 4)

index,_ = vq(data, centroids)
# plt.scatter(dfl, dfr)

# x = data.iloc[:, 0:1]

plot(data[index==0,0], data[index==0,1], 'or', data[index==1,0], data[index==1,1], 'ob', data[index==2,0], data[index==2,1], 'om', data[index == 3, 0], data[index == 3, 1], 'oc')
plot(centroids[:,0], centroids[:, 1], 'sg', markersize = 8)


show()
```

```python
#Exercise 2b(5 clusters)

centroids,_ = kmeans(data, 5)

index,_ = vq(data, centroids)

plot(data[index==0,0], data[index==0,1], 'or', data[index==1,0], data[index==1,1], 'ob', data[index==2,0], data[index==2,1], 'om', data[index == 3, 0], data[index == 3, 1], 'oc', data[index == 4, 0], data[index == 4, 1], 'ok')
plot(centroids[:,0], centroids[:, 1], 'sg', markersize = 8)


show()
```
